```python
import yt_dlp
import os

def download_youtube_audio(url, output_path='.', file_format='mp3'):
    """
    Downloads audio from a YouTube URL.

    Args:
        url (str): The YouTube video URL.
        output_path (str): The directory to save the audio file.
        file_format (str): The desired audio format (e.g., 'mp3', 'wav', 'm4a', 'opus').
    """
    # Create the output directory if it doesn't exist
    os.makedirs(output_path, exist_ok=True)

    ydl_opts = {
        'format': 'bestaudio/best',  # Download the best available audio format
        'outtmpl': os.path.join(output_path, '%(title)s.%(ext)s'), # Output file template
        'postprocessors': [{
            'key': 'FFmpegExtractAudio', # Use FFmpeg to extract audio
            'preferredcodec': file_format, # Preferred audio codec
            'preferredquality': '0',     # Best quality
        }],
        'ffmpeg_location': 'ffmpeg',  # Ensure ffmpeg is in your PATH or provide full path
        'verbose': False, # Set to True for more detailed output during download
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info_dict = ydl.extract_info(url, download=True)
            title = info_dict.get('title', 'unknown_title')
            print(f"Successfully downloaded: {title}.{file_format}")
            return os.path.join(output_path, f"{title}.{file_format}") # Return the downloaded file path
    except yt_dlp.DownloadError as e:
        print(f"Error downloading audio: {e}")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None


youtube_url = 'Ft-7at-gNTU'
output_dir = '../DALI-audio'

downloaded_file = download_youtube_audio(youtube_url, output_dir, file_format='wav')

if downloaded_file:
    print(f"Audio saved to: {downloaded_file}")
else:
    print("Audio download failed.")
```
