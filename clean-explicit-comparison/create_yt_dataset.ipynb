```python
import pysrt
import json

# === INPUT FILES ===
SRT_FILE = f"{title}.en.srt"           # YouTube auto-captions (SRT format)
WHISPER_JSON = f"{title}.json"         # Whisper output: list of word dicts

# === Parameters ===
DELAY = 0.05  # seconds to delay word start for better alignment
WORD_COUNT_TOLERANCE = 2  # max allowed word count difference before fallback

# === Load and enforce disjoint YouTube caption windows ===
subs = pysrt.open(SRT_FILE)
subs_data = []
for i, sub in enumerate(subs):
    start = sub.start.ordinal / 1000
    end = sub.end.ordinal / 1000
    # Clamp overlapping captions to be disjoint
    if i < len(subs) - 1:
        next_start = subs[i + 1].start.ordinal / 1000
        if end > next_start:
            end = next_start - 1e-3
    text = sub.text.strip()
    subs_data.append({"start": start, "end": end, "text": text, "words": []})

# === Load Whisper word-level JSON ===
with open(WHISPER_JSON, "r", encoding="utf-8") as f:
    whisper_words = json.load(f)

# === Assign Whisper words to captions using midpoint strategy with fallback ===
for word in whisper_words:
    ws = word.get("start")
    we = word.get("end")
    wtext = word.get("word").strip()
    if ws is None or we is None:
        continue

    midpoint = (ws + we) / 2
    ws_delayed = ws + DELAY

    assigned = False

    # Try assigning using midpoint if word count roughly matches
    for cap in subs_data:
        yt_word_count = len(cap["text"].split())
        wh_word_count = len(cap["words"])
        if cap["start"] <= midpoint < cap["end"]:
            if abs(yt_word_count - wh_word_count) <= WORD_COUNT_TOLERANCE:
                cap["words"].append(wtext)
                assigned = True
                break

    # Fallback: assign based on delayed start
    if not assigned:
        for cap in subs_data:
            if cap["start"] <= ws_delayed < cap["end"]:
                cap["words"].append(wtext)
                assigned = True
                break

    # Final fallback: assign to caption with maximum overlap
    if not assigned:
        best_cap = None
        best_overlap = 0.0
        for cap in subs_data:
            cs, ce = cap["start"], cap["end"]
            overlap = max(0.0, min(we, ce) - max(ws, cs))
            if overlap > best_overlap:
                best_overlap = overlap
                best_cap = cap
        if best_cap:
            best_cap["words"].append(wtext)

# === Display first few aligned lines ===
for c in subs_data[:10]:
    print(f"YT: {c['text']}")
    print(f"WH: {' '.join(c['words'])}")
    print(f"Time: {c['start']}–{c['end']}")
    print()
```
